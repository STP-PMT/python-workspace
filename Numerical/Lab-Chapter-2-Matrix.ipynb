```
#4.1
import numpy as np

A = np.array([[10,8,5,6],[6,23,14,12],[7,3,7,-9]])
B = np.array([[0,8],[1,12],[7,9],[2,6]])
C= np.matmul(A,B)

print('A : \n',A)
print('\nB : \n',B)
print('\nC = A x B \n',C)
```

```
#4.2
A =np.arange(6).reshape(3,2)
B = np.arange(6).reshape(3,2)
C = A+B

print('A : \n',A,'\n')
print('B : \n',B,'\n')
print('C : \n',C,'\n')
```

```
#4.3
A = np.random.randint(-5,5,size=(3,4))
B = np.random.randint(-2,2,size=(3,4))
C = A+B
D = A-B

print('A : \n',A,'\n')
print('B : \n',B,'\n')
print('C = A+B\n',C,'\n')
print('D = A-B\n',D,'\n')
```

```
#4.4
A = np.random.randint(1,10,size=(3,3))
B = np.random.randint(2,5,size=(3,2))

print('A : \n',A,'\n')
print('B : \n',B,'\n')
print('A x B\n',np.matmul(A,B),'\n')
print('transpose of A\n',A.transpose(),'\n')
print('Inverse of A\n',np.linalg.inv(A),'\n')
print('Determinant of A\n',np.linalg.det(A),'\n')
```

```
#4.5
A = np.array([[1,2,3,4],[8,7,6,5],[-2,-4,-6,-8]])

print('A : \n',A,'\n')
print('row#0:\t',A[0])
print('last row:\t',A[-1])
print('column#1:\t',A[:,1])
print('last col:\t',A[:,-1])
print('A[1][2]:\t',A[1][-3])
print('last element:\t',A[-1][-1],'\n')
print('row #0,1:\n',A[:2,:],'\n')
print('col #0,1:\n',A[:,1:3],'\n')
print('index of max value:\t',A.argmax())
print('max value:\t',A.max())
print('index of min value:\t',A.argmin())
print('min value:\t',A.min())
print('sum of each column:\t',A.sum(axis=0))
print('sum of each row:\t',A.sum(axis=1))
```
