```python
import numpy as np
import matplotlib.pyplot as plt 


lay1='y1 = 2x-5'
lay2='y2 = x-3'

x = np.linspace(-5,20)
y1=(2*x)-5
y2=x-3
plt.plot(x,y1,color='red',label =lay1)
plt.plot(x,y2,color='green',label =lay2)
plt.plot(2,-1,'bo')
plt.text(0,1,'(2,-1)',fontsize =15 )

plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10)
plt.figure()

plt.subplot(321)
plt.plot(x,np.cos(x),'r-')

plt.subplot(322)
plt.plot(x,np.sin(x),'g-')

plt.subplot(323)
plt.plot(x,np.tan(x),'b-')

plt.subplot(324)
plt.plot(x,-np.sin(x),'p-')

plt.subplot(325)
plt.plot(x,'y-')

plt.subplot(326)
plt.plot(x,-np.tan(x),'r--')

plt.show()
```
