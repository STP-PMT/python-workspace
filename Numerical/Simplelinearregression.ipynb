```python
import numpy as np 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error,r2_score 
import matplotlib.pyplot as plt 
import pandas as pd 
plt.style.use("seaborn")

x_train = np.array([1,2,2,3,4,6]).reshape((-1,1))
y_train = np.array([2,3,4,4,5,8])

model = LinearRegression()
model.fit(x_train,y_train)

b = model.coef_
a = model.intercept_

y_pred = model.predict(x_train)
r2 = r2_score(y_train,y_pred)
mse = mean_squared_error(y_train,y_pred)

x_new = np.array([10,12]).reshape((-1,1))
y_pred_new = model.predict(x_new)
print("R2 :%.2f"%(r2))
print("MSE :%.2f"%(mse))

plt.scatter(x_train,y_train,color = "green")
plt.plot(x_train,y_pred,color ="blue",label = "Y = %.2fX + %.2f"%(b,a))
plt.scatter(x_new,y_pred_new,color="red",marker="X")
plt.axis([0, 15, 0, 15])
plt.legend()
plt.show()
```
